```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "NanumBarunGothic"
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (10,7)
```

```python
데이터셋 = pd.read_excel( "/content/gdrive/MyDrive/스낵포_소스매체보고서.xlsx")
```

```python
데이터셋.info()
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
데이터셋_열 = len( 데이터셋.columns )
데이터셋_열_명칭리스트 = 데이터셋.columns.tolist()
데이터셋_열_숫자만 = 데이터셋.dtypes.tolist()
데이터셋_행 = len( 데이터셋 )
데이터셋_열_숫자만_잡아봐요 = [ i if 데이터셋_열_숫자만[i] != 'object' else None for i in range(데이터셋_열-1) ]
```

```python
for i in range( len(데이터셋_열_숫자만_잡아봐요) ) :
  if 데이터셋_열_숫자만_잡아봐요[i] != None :
    plt.title("데이터 선형관계 분석 (좌/우 상/하향 파악)")
    sns.regplot(데이터셋[데이터셋_열_명칭리스트[i]], 데이터셋[데이터셋_열_명칭리스트[-1]] ,marker = "*", color = "r", line_kws = {"color":"g","alpha":0.7,"lw":5} )
    plt.show()
```

```python
plt.title("데이터 상관관계 (데이터간 관계식 모델 구축 파악)")
sns.heatmap(데이터셋.corr(), linewidths = 0.1, vmax = 0.5, fmt = ".2f", cmap = plt.cm.gist_heat, linecolor = "white", annot = True)
plt.show()
```
